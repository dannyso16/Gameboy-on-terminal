# 1.ANSIエスケープコード

## 基本

※Colaboratoryは、ANSIエスケープコードが動作しないので、縦に並んで出力されます。

```
import time

spin_characters = ["( ˘ω˘  )",
                   "(   ˘ω )",
                   "(      ˘)",
                   "(˘      )",
                   "(ω˘    )",
                   "(˘ω˘   )"
                   ]
cycle = len(spin_characters)

for i in range(cycle*5):
    i %= cycle
    s = spin_characters[i]
    print(s, end="")
    time.sleep(0.2)
    print("\033[1A")

# 後処理
print("\033[2K")
```
